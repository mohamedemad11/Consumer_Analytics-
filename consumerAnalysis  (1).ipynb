```python
data.dropna(axis=0,inplace=True)
```

```python
data['user_age']=2021-data['Year_Birth']
```

```python
data.head()
```

```python
data.info()
```

```python
data['Dt_Customer']=pd.to_datetime(data['Dt_Customer'])
```

```python
data['year_of_join']=data['Dt_Customer'].dt.year
```

```python
data.head()
```

```python
data['join_years']=2021-data['year_of_join']
```

```python
data.head()
```

# what are the percentage of our customers education ?

```python
fig=plt.figure(figsize=(20,10))
plt.title('% of our customers education',fontsize=20)
data.Education.value_counts(normalize=True).plot(kind='pie',autopct='%1.1f%%')
plt.show()
fig.savefig("percentage_of_our_customer_education.png", bbox_inches='tight')
```


### The majority of our consumers are university graduated , the rest of them are phd and master holders 

```python
plt.figure(figsize=(20,10))
sns.boxplot(data.user_age)
```

```python
data=data[data['user_age']<=100]
```

# what is the distribution of our consumers ages ?


```python
fig=plt.figure(figsize=(20,10))
sns.boxplot(data.user_age)
plt.title('Distriution of our ages ')
plt.show()
fig.savefig("Distribution_of_our_customer_ages.png", bbox_inches='tight')
```

### Our Niche consumers are people from 43 to 63 years old 

```python
fig=plt.figure(figsize=(20,10))
sns.distplot(data.user_age)
plt.xticks(range(0,100,5),fontsize=15)
plt.title('distribution of our ages ')
plt.show()
fig.savefig("distribution_of_our_ccconsumers_ages.png", bbox_inches='tight')
```

```python
data.head()
```

# what is the percentage of our marital status 

```python
fig=plt.figure(figsize=(20,10))
plt.title('Our User Marital Status ')
data.Marital_Status.value_counts(normalize=True).plot(kind='pie',autopct='%1.1f%%')
plt.show()
fig.savefig("percentage_of_our_customer_Marital_status.png", bbox_inches='tight')
```


### our niche users are Married and living together , so we should focus on prodcuts to engaged people 

```python
plt.figure(figsize=(20,10))
sns.distplot(data.join_years)
```

# what is the percentage of our customer loyality years ?


```python
fig=plt.figure(figsize=(20,10))
plt.title('Our User Loyalty years ')
data.join_years.value_counts(normalize=True).plot(kind='pie',autopct='%1.1f%%')
plt.show()
fig.savefig("percentage_of_our_customer_loyalty_years.png", bbox_inches='tight')
```
